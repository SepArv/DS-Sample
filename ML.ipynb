```python
import yfinance as yf

msft = yf.Ticker("MSFT")

# get all stock info
msft.info

# get historical market data
hist = msft.history(period="2y")
```

```python
hist['SMA10'] = hist['Close'].rolling(10).mean()
hist['Dec_SMA10'] = hist.SMA10 < hist.SMA10.shift()
hist['SMA30'] = hist['Close'].rolling(30).mean()
hist['Dec_SMA30'] = hist.SMA30 < hist.SMA30.shift()
hist.dropna(inplace=True)
```

```python
hist
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(hist['SMA10'], label = 'SMA10')
ax.plot(hist['SMA30'], label = 'SMA30')
ax.set_xlabel = ['Eur']
ax.set_ylabel = ['Date']
```
